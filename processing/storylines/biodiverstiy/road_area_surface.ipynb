```python
import contextily as ctx
import geopandas as gpd
import pandas as pd
import fiona
import os
```

```python
data_dir = os.path.join(os.getcwd(), 'data')

gdb_path = os.path.join(data_dir, 'Lebensraumkarte_v1_1_VD_20241025.gdb')
layers = fiona.listlayers(gdb_path)
areas = gpd.read_file(gdb_path, layer=layers[0], columns=['TypoCH_NUM', 'TypoCH', 'POLYID', 'geometry'])

roads_path = os.path.join(data_dir, 'lausanne_drive_edges.gpkg')
roads = gpd.read_file(roads_path, columns=['geometry'])

roads_areas_filename = os.path.join(data_dir, "roads_areas_unprocessed.gpkg")
roads_areas = gpd.read_file(roads_areas_filename)
```

```python
roads_areas["area"] = roads_areas["geometry"].area
```

```python
roads_areas.drop(columns=['geometry', 'TypoCH_NUM', 'POLYID', 'TypoCH', 'TypoCH_v2', 'cluster_v2', 'side'], inplace=True)
roads_areas['road_geometry'] = gpd.GeoSeries.from_wkt(roads_areas['road_geometry'])
roads_areas = roads_areas.rename(columns={'road_geometry': 'geometry'})
roads_areas = roads_areas.set_geometry('geometry')
```

```python
roads_areas = roads_areas.dissolve(by='road_id', aggfunc='sum')
```
